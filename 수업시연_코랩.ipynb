```python
# '단어'가 출현하는 횟수를 측정하는 코드
scriptSample.count('단어')
```

```python
# 공백을 기준으로 단어를 구분하여 listStory에 넣음
listStory = scriptSample.split()
len(listStory) # 단어가 몇개인지 세는 코드
```

```python
# 처음 3개의 단어를 출력하는 코드
listStory[:3]
# 전체를 출력하는 코드
listStory
```

```python
# 단어 별로 출현횟수를 세는 코드
dictStory = {}
for item in listStory:
    if item in dictStory:
        dictStory[item] += 1
    else:
        dictStory[item] = 1
dictStory
```

```python
import pandas as pd
df = pd.DataFrame(dictStory, index=[0]).T
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 빈도수가 2 이상인 단어만 추출, 빈도수 기준을 바꾸려면 5를 다른 숫자로 바꾸면 된다.
df_filtered = df[df[0] >= 2]
# 오름차로 설정, 빈도가 가장 높은 단어부터 나타남.
df_filtered = df_filtered.sort_values(by=0, ascending=False)
```

```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rc('font', family='NanumGothic')  # 한글 폰트
plt.rc('axes', unicode_minus=False)  # 음수 부호 표시 설정

# 바 그래프 그리기
if not df_filtered.empty:  # df_filtered가 비어 있지 않은 경우에만
    plt.figure(figsize=(10, 6))  # 크기를 (10, 6)으로 설정
    plt.bar(df_filtered.index, df_filtered[0])
    plt.xticks(rotation=90)  # 가로축의 이름을 90도 회전
    plt.xlabel('단어')  # 가로축의 이름
    plt.ylabel('빈도수')  # 세로축의 이름
    plt.title('단어 빈도수')  # 제목
    plt.savefig('bar_graph.png') # "bar_graph.png로 저장"
    plt.show()
else:
    print("단어 수가 1개 이하입니다. 표시할 데이터가 없습니다.")  # 단어 데이터프레임에 단어가 없을 경우
```
